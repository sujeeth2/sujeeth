## fit a linear regression model

```python
import statsmodels.formula.api as smf
model1 = smf.ols("sunday~daily",data = data1).fit()
```

```python
model1.summary()
```

```python
x = data1["daily"].values
y = data1["sunday"].values
plt.scatter(x, y, color ="m", marker = "o", s=30)
b0 = 13.84
b1 = 1.33

y_hat = b0 +b1*x

plt.plot(x, y_hat, color ="g")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Plot the linear regression line using seaborn regplot() method
sns.regplot(x="daily", y="sunday", data=data1)
plt.xlim([0,1250])
plt.show()
```

```python
#Predict sunday sales lfor 200 and 300 and 1500 daily circulation
newdata=pd.Series([200,300,1500])
```

```python
data_pred=pd.DataFrame(newdata, columns=['daily'])
data_pred
```

```python
model1.predict(data_pred)
```

```python
pred = model1.predict(data1["daily"])
pred
```

```python
data1["Y_hat"] = pred
data1
```

```python
data1["residuals"]= data1["sunday"]-data1["Y_hat"]
data1
```

```python
mse = np.mean((data1["daily"]-data1["Y_hat"])**2)
rmse = np.sqrt(mse)
print("MSE: ",mse)
print("RMSE: ",rmse)
```

```python
mae = np.mean(np.abs(data1["daily"]-data1["Y_hat"]))
mae
```

## Checking the model rediuals fo scatter ploy ( for homoscedasticity) and normality
- check the scater plot for residuals Vs Y_hat
- check the Q-Q plot for residuals distribution
- check the kde hist plot for residuals

```python
plt.scatter(data1["Y_hat"], data1["residuals"])
```

## Observation:
- There appears to be no trend and the residuals are randomly placed around the zero error line
- Hence the assupmtion of homescedaticity (constant variance in residuals) is satisfied

```python
# plot the Q-Q plot (to check the normality of residuals)
import statsmodels.api as sm
sm.qqplot(data1["residuals"], line='45', fit=True)
plt.show()
```

```python
# plot the kde distribution for residuals
sns.histplot(data1["residuals"], kde=True)
```

## observation:
- The data points are seen to closely follow the reference line of normality
- Hence the residuals are approximately normally distributed a also can be seen from the kde distribution
- All the assumptions of simple linear regression are verified and found to be vaild in tis case
